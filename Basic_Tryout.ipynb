```python
from langchain_ollama.chat_models import ChatOllama
from langchain.chat_models import init_chat_model
from langgraph.graph import StateGraph, START, END, MessagesState
from langgraph.graph.message import add_messages
from typing import TypedDict, Annotated
```

```python
#llm = ChatOllama(model="mistral:latest", temperature=0.1)
llm = init_chat_model(model_provider="ollama", model="mistral:latest", temperature=0.1)
```

```python
# class MessageState(TypedDict):
#     messages: Annotated[list, add_messages]

class MessageState(MessagesState):
    pass
```

```python
def chat_bot(msg_state: MessageState):
    return {"messages": [llm.invoke(msg_state["messages"])]}
```

```python
graph_builder = StateGraph(MessageState)
graph_builder.add_node("chat_bot",chat_bot)
graph_builder.add_edge(START, "chat_bot")
graph_builder.add_edge("chat_bot",END)
workflow = graph_builder.compile()
#workflow
```

```python
user_input:str = str(input("User: "))
response = workflow.invoke(
    {"messages":[{"role":"user","content":user_input}]}
)
#print(response)
print(response["messages"])
for msg in response["messages"]:
    msg.pretty_print()
```
